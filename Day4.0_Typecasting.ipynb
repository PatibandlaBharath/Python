# Type Casting

*Converting from one data type to another*

#### *Converting int ->float,bool,str,comp*

```python
a=10
type(a)
```

```python
# Converting int to float
float(a)
```

```python
# Converting int to boolean
bool(a)
```

```python
# Converting int to String
str(a)
```

```python
# Converting int to complex
complex(a)
```

#### *Converting float ->int,bool,str,comp*

```python
b=13.45
type(b)
```

```python
# Converting float to int
int(b)
```

```python
# Converting float to string
str(b)
```

```python
# Converting float to boolean
bool(b)
```

```python
# Converting float to complex
complex(b)
```

#### *Converting String ->int,float,bool,comp*

*Point to note---String can't be converted to anything except* ***bool***

```python
# int(n) --- not possible->error
# float(n) --- not possible->error
# complex(n) --- not possible->error
```

```python
# Converting string to boolean
# If the variable contains a string,the o/p will be true
n='Zade'
bool(n)
```

```python
# If the variable has no string,the o/p will be false
m=""
bool(m)
```

#### *Converting Complex ->int,float,bool,string*

*Point to note---Complex can't be converted to* ***int, float***

```python
comp=5+8j
type(comp)
```

```python
# int(comp) --- not possible->error
# float(comp) --- not possible->error
```

```python
# Converting complex to string
str(comp)
```

```python
# Converting complex to boolean
bool(comp)
```
