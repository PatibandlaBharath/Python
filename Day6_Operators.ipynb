#### Ternary operator

**x =first value if condition else second value.**

*If the condition is True, it returns firstvalue.
else, returns second value.*

```python
a=10
b=20
x=30 if a>b else 40
x
```

```python
# Taking input from keyboard and print minimum value
a=int(input("enter a"))
b=int(input("enter b"))
min=a if a<b else b
min
```

```python
# Taking input from keyboard and print minimum value for 3 variables
x=int(input("Enter x"))
y=int(input("Enter y"))
z=int(input("Enter z"))
min=x if x<y and x<z else y if y<x and y<z else z
print("Minimum value is:",min)
```

```python
# Taking input from keyboard and print maximum value for 3 variables
x=int(input("Enter x"))
y=int(input("Enter y"))
z=int(input("Enter z"))
max=x if x>y and x>z else y if y>x and y>z else z
print("Maximum value is:",max)
```

### Special Operators

*There are two special operators in python.*
- Identity Operator
- Membership Operator

***Identity Operator :*** We use identity operators for **addresss comparision**.There are two identity operators available.
1. is
2. is not

```python
a=10
b=10
a is b
```

```python
id(a)
```

```python
id(b)
```

```python
a is not b
```

```python
a=5+8j
id(a)
```

```python
b=5+8j
id(b)
```

```python
# values are same but obj id is not same
a is b
```

```python
a="Ronan"
b="Ronan"
id(a)
```

```python
id(b)
```

```python
a is b
```

```python
list1=["one","two","three"]
```

```python
list2=["one","two","three"]
```

```python
id(list1)
```

```python
id(list2)
```

```python
# values are same but obj id is not same
list1 is list2
```

***Membership operator :***
1. in - return true if the given object present in the sppecified collection
2. not - in return true if the given object present not present in the sppecified collection

```python
x="Learning python is very easy"
'L' in x
```

```python
'z' in x
```

```python
'z' not in x
```

```python
list1=["one","two","three","four"]
'four' in list1
```
