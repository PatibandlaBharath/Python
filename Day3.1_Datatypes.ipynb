# Datatypes

*They represent the type of data present inside a variable*

*In python it is not required to specify the type of the data explicitly*

*Based on the value provided,the type will be assigned automatically.*

*So it is dynamically typed language.*

*type() ->It is an in-built func in python to show the data type of a value*

```python
tot=10
type(tot)
```

### No. of Data Types(Built-in)

1. *int ->numbers(...,-2,-1,0,1,2,...)*
2. *float ->decimal value*
3. *complex ->real+imaginary values(6+8j)*
4. *str ->character or group of characters*
5. *bool ->True/False*

```python
# integer datatype
a=10
print(type(a))
# float datatype
b=12.23232
print(type(b))
```

```python
# Complex datatype
x=6+8j
print(type(x))
print(x.real)
print(x.imag)
```

```python
# String datatype
n='Landon'
m="Jeremy"
p='''aaron 
warner
anderson'''
print(n)
print(m)
print(p)
# if we use just p instead of print(p) it'll give "aaron \nwarner\nanderson"
```

```python
# Boolean datatype
l=True
print(type(l))
g=10
h=20
i=10
print(g is h)
print(g is i)
```

```python
print(id(g))
print(id(h))
print(id(i))
```

### Representation of int in multiple format

1. *Binary(base:2)*
2. *Decimal(base:10)*
3. *Octal(base:8)*
4. *Hexadecimal(base:16)*

```python
binary=0b1111
decimal=1000
octal=0o456
hexadecimal=0xA69
```

```python
binary,decimal,octal,hexadecimal
```

##### Conv dec to all

```python
a=3334
bin(a),oct(a),hex(a)
```

##### Oct to all
