## While Loop

if we want to execute a group of statements itteratively until some condition false,then go for while loop.

```python
# To print number from 1 to 10 in while loop
x=1
while x<=10:
    print(x)
    x=x+1
```

```python
# Display the sum of first n number
n=int(input("enter a number"))
sum=0
i=1
while i<=n:
    sum=sum+i
    i=i+1
    print("the sum of first",n,"number",sum)
```

# Transefer Statements

break

```python
for i in range(10):
    if i==7:
        print("processing is enough,plz break")
        break
    print(i)
```

```python
cart=[100,200,400,600,700,900]
for item in cart:
    if item>500:
        print("you get 20% discount on this order")
        break
    print(item) 
```

## Continue

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
cart=[100,200,400,600,700,900]
for item in cart:
    if item>=500:
        print("We can not process this item",item)
        continue
    print(item)
```
