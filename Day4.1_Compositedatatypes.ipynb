# Composite Data Types

```python
b=[10,20,30,40,50]
type(b)
```

### Bytes Data types

```python
x=bytes(b)
type(x)
```

```python
x[0]
```

```python
for i in x:
    print(i)
```

*Points to be noted:*
1. *It does not support item assignment because it is immutable.*
2. *Value-once defined, cannot be modified.*

```python
# x[1]=16 ---error -> 'bytes' object does not support item assignment*
```

### Bytesarray Datatype

*It is same as bytes data type and is mutable*

```python
y=[10,20,30,40,50]
ba=bytearray(y)
type(ba)
```

```python
for i in ba:
    print(i)
```

```python
ba[2]
```

```python
#Changing the value of ba[2]
ba[2]=43 
```

```python
for i in ba:
    print(i)
```

### List

*A List is a collection of hetrogeneous elements(can be int,float,str or anything)*

*If we want to represent a group of data as a single entity where:*
1. *Insertion order is preserved*
2. *hetrogeneous elements are allowed*
3. *Duplicates are allowed*
4. *Growable in nature*
5. *It's mutable*
6. *Value enclosed in [] and seperated by ,*

```python
list1=[10,20.4,'aiden',7+8j,True,10]
```

```python
list1[0]
```

```python
list1[3]
```

```python
# Changing the value of list1[5]
list1[5]=35
```

```python
# Adding a value to the list using append()
list1.append(124)
```

```python
list1
```

```python
# Removing a value from the list using remove()
list1.remove(124)
```
