### Rules to define identifiers in python

##### 1. *Only alphabets symbols are allowed(upper & lower case)* 
-> *A-Z , a-z*
##### 2. *Digits are allowed* 
-> *0-9*
##### 3. *Underscore symbol is allowed*
##### 4. *$ is not allowed*

```python
cash=2000
```

```python
# ca$h=60 -> will show syntax error
```
