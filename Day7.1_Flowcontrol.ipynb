# Flow Control

*Describe the order in which the statements will be executed.*

***Types of Control flow***
1. *Conditional Statements -> if, else, elif*
2. *Iterative Statements -> for, while*
3. *Transfer Statements -> break, continue*

### Conditional Statements

#### if

```python
# Single statement
name=input("Enter a name")
if name=="Levi":print("hi",name)
```

```python
# Multiple statements
if name=="Levi":
    print("Hello {0}".format(name))
    print("Hi {0}".format(name))
```

#### if-else

```python
name=(input("Enter the name of guest"))
if name=="Astrid":
    print("Hi,",name," Welcome to our hotel")
else:
    print("Sorry,",name,"We don't have a reservation on your name")
```

#### if-elif-else

```python
brand=input("What is your favourite brand:")
if brand=="hermes":
    print("{0} mostly sell luxury bags".format(brand))
elif brand=="Van Cleef":
    print("{0} mostly sell jewlery".format(brand))
else:
    print("No discount")
```

```python
# finding whether the number belongs to 1-100 or not
n=int(input("Enter a number b/w 1-100:"))
if 1<n<100 :
    print("The number belongs to 1-100 set")
else:
    print("The number doesn't belong in set 1-100 ")
```

```python
# finding the biggest/smallest among 3 values
a,b,c=(input("Enter 3 values:")).split()
if a>b and a>c:
    print("The biggest among three is {0}".format(a))
elif b>a and b>c:
    print("The biggest among three is {0}".format(b))
else:
    print("The biggest among three is {0}".format(c))
```

```python
# Even/Odd
a=int(input("enter a value"))
if a%2==0:
    print("The number is even")
else:
    print("The number is odd")
```
